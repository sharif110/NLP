```python
from tabulate import tabulate

corpora_info = [
    (text1, "Text1"),
    (text2, "Text2"),
    (text3, "Text3"),
    (text4, "Text4"),
    (text5, "Text5"),
    (text6, "Text6"),
    (text7, "Text7"),
    (text8, "Text8"),
    (text9, "Text9")
]

# Initialize a list to store rows of the table
table_rows = []

# Fill in the table
for corpus, corpus_name in corpora_info:
    corpus_length = len(corpus)
    unique_words = len(set(corpus))
    lexical_richness = unique_words / corpus_length if corpus_length > 0 else 0
    
    # Append a row to the table
    table_rows.append([corpus_name, corpus.name, corpus_length, unique_words, lexical_richness])

# Define headers for the table
headers = ["Corpus", "Corpus Name", "Corpus Length", "Unique Words", "Lexical Richness"]

# Display the table using tabulate
table_str = tabulate(table_rows, headers=headers, tablefmt="grid")

# Print the table
print(table_str)
```

```python
import math
from nltk import FreqDist
from nltk.corpus import stopwords
```

```python
def TF(corpus, token):
    return corpus.count(token) / len(corpus) * 100
```

```python
def LOGTF(corpus, token):
    return math.log(corpus.count(token) + 1, 10)
```

```python
def IDF(corpus, token):
    return math.log(len(corpora_info) / corpus.count(token), 10)
```

```python
# Obtain the 10 most common words in text1 using the Frequency Distribution
fdist1 = FreqDist(text1)
common_words = [word for word, _ in fdist1.most_common(10)]
```

```python
# Define the words to analyze
words_to_analyze = ["monster", "evil", "devil", "the"]
```

```python
# Initialize a list to store rows of the table
table_rows = []

# Fill in the table
for word in words_to_analyze:
    table_rows.append([
        word,
        TF(text1, word),
        LOGTF(text1, word),
        IDF(text1, word),
        fdist1[word] if word in fdist1 else ''
    ])
```

```python
# Additional information for Common Words (top 3)
common_words_top3 = fdist1.most_common(3)
for i, (word, count) in enumerate(common_words_top3, 1):
    table_rows.append(["Common word " + str(i), '', '', '', count])
```

```python
# Define headers for the table
headers = ["Tokens", "TF()", "LOGTF()", "IDF()", "Common Words"]

# Display the table using tabulate
table_str = tabulate(table_rows, headers=headers, tablefmt="grid", numalign="right")

# Print the table
print(table_str)
```

```python
nltk.download('punkt')
text = "NLTK is a powerful library for natural language processing."
words = nltk.word_tokenize(text)
sentences = nltk.sent_tokenize(text)
print(words)
print(sentences)
```

```python
# To to the text, assign parts of speech tags to the sentencs:
nltk.download('averaged_perceptron_tagger')
tags = nltk.pos_tag(words)
print(tags)
```

```python
# Remove some of the stop words like is, a, etc. in the sentence:
nltk.download('stopwords')
from nltk.corpus import stopwords
filtered_words = [word for word in words if word.lower() not in
stopwords.words('english')]
print(filtered_words)
```
