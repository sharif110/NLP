```python
# Function to filter words based on conditions
def filter_words(words, condition):
    return [w for w in words if condition(w)]
```

```python
# Extracting unique sorted words without punctuations for text1, text2, and text3
words_text1 = sorted(set(text1))[280:]
words_text2 = sorted(set(text2))
words_text3 = sorted(set(text3))
```

```python
# filtering for text1
long_words_text1 = filter_words(words_text1, lambda w: len(w) > 16)
high_freq_text1 = filter_words(words_text1, lambda w: text1.count(w) > 500)
ed_words_text1 = filter_words(words_text1, lambda w: w.endswith('ed'))

# filtering for text2
long_words_text2 = filter_words(words_text2, lambda w: len(w) > 16)
high_freq_text2 = filter_words(words_text2, lambda w: text2.count(w) > 500)
ed_words_text2 = filter_words(words_text2, lambda w: w.endswith('ed'))

# filtering for text3
long_words_text3 = filter_words(words_text3, lambda w: len(w) > 16)
high_freq_text3 = filter_words(words_text3, lambda w: text3.count(w) > 500)
ed_words_text3 = filter_words(words_text3, lambda w: w.endswith('ed'))
```

```python
# Creating bigram and trigram models for text1
bigrams_text1 = list(bigrams(text1))
trigrams_text1 = list(trigrams(text1))
trigrams_text1_collocations = TrigramCollocationFinder.from_words(text1).nbest(TrigramAssocMeasures().pmi, 5)

# Creating bigram and trigram models for text2
bigrams_text2 = list(bigrams(text2))
trigrams_text2 = list(trigrams(text2))
trigrams_text2_collocations = TrigramCollocationFinder.from_words(text2).nbest(TrigramAssocMeasures().pmi, 5)

# Creating bigram and trigram models for text3
bigrams_text3 = list(bigrams(text3))
trigrams_text3 = list(trigrams(text3))
trigrams_text3_collocations = TrigramCollocationFinder.from_words(text3).nbest(TrigramAssocMeasures().pmi, 5)
```

```python
# Filling the table
table_data = [
    (len(set(bigrams_text1)), len(set(trigrams_text1)), len(long_words_text1), len(high_freq_text1), len(ed_words_text1)),
    (len(set(bigrams_text2)), len(set(trigrams_text2)), len(long_words_text2), len(high_freq_text2), len(ed_words_text2)),
    (len(set(bigrams_text3)), len(set(trigrams_text3)), len(long_words_text3), len(high_freq_text3), len(ed_words_text3))
]
```

```python
import os
from nltk.util import ngrams
```

```python
# Specify the location of your corpus file
corpus_path = os.path.expanduser('Lorem.txt')
```

```python
# Load the data in raw and text formats
psh_raw = nltk.data.load(corpus_path, format='raw')
psh_txt = nltk.data.load(corpus_path, format='text')
```

```python
# Display a snippet of the loaded data
print("\nSnippet of Raw Data:")
print(psh_raw[:100])
```

```python
# Tokenize and get bigrams and trigrams
words = nltk.word_tokenize(psh_txt)
psh_bigrams = list(ngrams(words, 2))
psh_trigrams = list(ngrams(words, 3))
```

```python
# Display the first 10 bigrams and trigrams
print("\nFirst 10 Bigrams:", psh_bigrams[:10])
print("\nFirst 10 Trigrams:", psh_trigrams[:10])
```

# Task 3: Generating Random Text with Bigrams


```python
cfd = nltk.ConditionalFreqDist(psh_bigrams)
```

```python
def generate_model(cfdist, word, num):
    for i in range(num):
        print(word, end=' ')
        word = cfdist[word].max()
```

```python
# Generate 10 words
generate_model(cfd, 'text', 10)
```
